```python
import numpy as np
import matplotlib.pyplot as plt
```

## Array advanced Operation

### 1. Example using Business Data

```python
# Data structure: [restaurant_id, 2022, 2023, 2024, 2025]
sales=np.array([[1, 100, 200, 300, 400],     #R1
                [2, 200, 400, 600, 800],     #R2
                [3, 300, 600, 900, 1200],    #R3
                [4, 400, 800, 1200, 1600]])  #R4

print("Sales data shape: ", sales.shape)
print("1st 3 restaurant:\n ", sales[:3], "\n")

yt=np.sum(sales[:,1:],axis=0) ## Sums column elements 
print("Yearly total: ", yt)

min=np.min(sales[:,1:],axis=1)
print("Minimum sales per restaurant: ", min)

max=np.max(sales[:,1:],axis=1)
print("Maximum sales per restaurant: ", max)

avg=np.mean(sales[:,1:], axis=1)
print("Average sales: ",avg)
```

### 2.Graph plot using Matplotlib

```python
cumsum = np.cumsum(sales[:, 1:], axis=1)       # Compute cumulative sum of sales per restaurant across years
print("Cumulative sum: \n", cumsum)             # Print the cumulative sales matrix

year = [2022, 2023, 2024, 2025]                 # Define x-axis values (years)

plt.figure(figsize=(10, 6))                     # Create a new figure of size 10x6 inches

plt.plot(year, np.mean(cumsum, axis=0))         # Plot average cumulative sales across all restaurants

plt.title("Average cumulative sales across all restaurants")  # Set plot title
plt.xlabel("Year")                              # Label x-axis
plt.ylabel("Sales")                             # Label y-axis
plt.grid(True)                                  # Add grid lines to the plot

plt.xticks(year)                                # Force x-axis to display only the given years
plt.show()                                      # Show the final plot
```

### Vector Addition

```python
import numpy as np   # Import NumPy library for numerical operations

# Define two vectors
v1 = np.array([1, 2, 3])
v2 = np.array([4, 5, 6])

# Vector addition (element-wise)
v3 = v1 + v2       

# Vector multiplication (element-wise, not dot product)
v4 = v1 * v2       

# Dot product of two vectors
v5 = np.dot(v1, v2)  

# Magnitude (length) of each vector using Euclidean norm
v1_mag = np.linalg.norm(v1)
v2_mag = np.linalg.norm(v2)

# Compute angle between vectors in radians using formula:
# cos(θ) = (v1 · v2) / (||v1|| * ||v2||)
ang = np.arccos(v5 / (v1_mag * v2_mag))

# Convert angle from radians to degrees for better interpretation
deg = np.degrees(ang)

# Print results
print("Vector Sum: ", v3)
print("Vector multiplication: ", v4)
print("Vector dot product :", v5)
print("Vector 1 magnitude : ", v1_mag)
print("Vector 2 magnitude : ", v2_mag)
print("Angle between them in radians:", ang)
print("Angle between them in degrees :", deg)


```
